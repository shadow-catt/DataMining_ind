# 2. Data Preprocessing

```python
#    buying       v-high, high, med, low
#    maint        v-high, high, med, low
#    doors        2, 3, 4, 5-more
#    persons      2, 4, more
#    lug_boot     small, med, big
#    safety       low, med, high

# Data preprocessing
def classify(x, cls):
        return list(cls).index(x)
def preprocess(df, proportion=.8, if_seperate=True,if_reverse=False):
    rows=df.shape[0]
    col = df.columns.values.tolist()
    # mix the number sequence of rows
    index_list=list(range(rows))
    np.random.shuffle(index_list)
    
    tmp_df=df.copy(deep=True)
    
    #print(mix_df)
    # get the unique coloums
    
    buying=['low','med','high','vhigh']
    maint=['low','med','high','vhigh']
    doors=['2','3','4','5more']
    persons=['2','4','more']
    lug_boot=['small','med', 'big']
    safety=['low','med','high' ]
    if if_reverse:
        evaluation=['unacc', 'acc']
    else:
        evaluation=['acc', 'unacc']
    
#     list_col=['buying','maint','doors','persons','lug_boot','safety','evaluation']
    list_col=list(tmp_df)
#     for each in list_col:
#         ucs.append(each)
    # replace all the df to number
    coloums=0
    for c in list_col:
        true_var=eval(c)
        list_num=list(range(len(true_var)))
        tmp_df[c]=tmp_df[c].replace(true_var,list_num)
    
    
    # store the df in random line
    mix_df = tmp_df.iloc[index_list]
    
    p_train = int(np.floor(rows * proportion))
    training = mix_df.iloc[:p_train]
    testing = mix_df.iloc[p_train:]
    if if_seperate:
        return training, testing
    else:
        # eturns data that has been processed but not out of order
        return tmp_df
# separate the dataframe into 4 pieces

def separate_df(df, proportion=.8,if_reverse=False):
    training, testing=preprocess(df, proportion,True,if_reverse)
    train_x=training.iloc[:,:-1]
    train_y=training.iloc[:,-1]
    test_x=testing.iloc[:,:-1]
    test_y=testing.iloc[:,-1]
    return train_x,train_y,test_x,test_y

# restore the orginal data
def restore(rev):
    buying=['low','med','high','vhigh']
    maint=['low','med','high','vhigh']
    doors=['2','3','4','5more']
    persons=['2','4','more']
    lug_boot=['small','med', 'big']
    safety=['low','med','high' ]
    evaluation=['unacc', 'acc']
    predict=['unacc', 'acc']
        
    list_col=list(rev)
    for coll in list_col:
        true_var=eval(coll)
        list_num=list(range(len(true_var)))
#         print(list_num)
#         print(true_var)
        rev[coll]=rev[coll].replace(list_num,true_var)
    return rev

# round the array to what we want
def round_list(float_list,round_num):
    # let it become array
    mid_np = np.array(float_list)                    
    # round it
    mid_np_2f = np.round(mid_np,round_num)
    # change the float_list
    list_new = list(mid_np_2f)                    
    return list_new
```

# 3. Index function

```python
# the output is accur,precision,recall,F_score
def cor_percentage(pred,actual,B2=1,heat=False):   #input should be pd.series
    try:
        # solve the problem caused by transer to series from array
        pred=pd.Series(pred.tolist())
        actual=pd.Series(actual.tolist())
        # get the row of row number of dataframe
        df_row_num=len(actual)

        # get the accuracy
        pred_actual=abs(pred-actual)
        same_num=df_row_num-sum(pred_actual)
        accur=same_num/df_row_num

        pred_sum=sum(pred)
        actual_sum=sum(actual)
        
        # get the Precision and recall
        add_list=np.array(pred+actual)
        same1_count=len(add_list[add_list > 1])
        same0_count=len(add_list[add_list < 0.5])

        precision=(same1_count)/(pred_sum)
        recall=(same1_count)/(actual_sum)
        
        F_score=((1+B2)*precision*recall)/(recall+precision*B2)
        
#         if heat:
#             tmp=pd.DataFrame({"PT":[same1_count, pred_sum-same1_count], "PN":[actual_sum-same1_count, same0_count]},index=["AT", "AN"])
#             heatmap(df)
        
        return accur,precision,recall,F_score
    except:
        return 0,0,0,0

# get the heap map of the index
# def heatmap(df):
#     plt.figure(figsize = (6, 5))
#     sns.set(font='simhei', font_scale=1.5)
#     sns.heatmap(df,fmt="d")
#     plt.show()
```

# 4. KNN

```python
# KNN for certain k
def judge_knn(training_mod,k,train_x,train_y,test_x,test_y=0):
    finial_list=[]
    test_x_height=test_x.shape[0]
    for i in range(test_x_height):
        
        # sub the i line and take abs for the dataframe
        
        # euclidean distance
        abs_subp=abs(train_x-test_x.iloc[i])
        #sum by line
        distance_list=((abs_subp != 0).astype(int).sum(axis=1))
    
    
        #store the distance and the result into result df
        result_df=pd.concat([distance_list,train_y],axis=1)
        # change the dataframe to array
        result_list=result_df.values
        
        # return the min distance with it result
        closest=result_list[sorted(np.argsort(result_list.T[0])[:k])]
        counts = np.bincount(closest[0])
        
        #get the counts max result
        type_w = np.argmax(counts)                                       
        
        # put the result for the pridiction line
        finial_list.append(type_w)
    if(training_mod==1):
        pr = test_x.assign(evaluation=pd.Series(test_y).values)
        # insert all the result into the predict dataframe
        pr2 = pr.assign(predict=pd.Series(finial_list).values)
    else:
        pr2 =  test_x.assign(predict=pd.Series(finial_list).values)
    
    return pr2
```

## c) Make improvement for changing the method to judge distance

Since there are many ways to calculate the distance and many K for the using, we can use some method to view their effectiveness. Thus, I rewrite the method and add some other ways to make pridict to the result.

```python
# KNN for certain k
def judge_knn_m(training_mod,method,k,train_x,train_y,test_x,test_y=0):
    tf=train_x.assign(evaluation=pd.Series(train_y).values)
    finial_list=[]
    test_x_height=test_x.shape[0]
    for i in range(test_x_height):
        
        # sub the i line and take abs for the dataframe
        
        if method==1:
            distance_list=calculate_manhattan(train_x,test_x,i)
        elif method==2:
            distance_list=calculate_euclidean(train_x,test_x,i)
        else:
            distance_list=calculate_different(train_x,test_x,i)
    
        #store the distance and the result into result df
        result_df=pd.concat([distance_list,tf.iloc[:,-1]],axis=1)
        # change the dataframe to array
        result_list=result_df.values
        
        # return the min distance with it result
        closest=result_list[sorted(np.argsort(result_list.T[0])[:k])]
        counts = np.bincount(closest[0])
        
        #get the counts max result
        type_w = np.argmax(counts)                                       
        
        # put the result for the pridiction line
        finial_list.append(type_w)
    if(training_mod==1):
        try:
            pr = test_x.assign(evaluation=pd.Series(test_y).values)
            # insert all the result into the predict dataframe
            pr2 = pr.assign(predict=pd.Series(finial_list).values)
        except:
            print("You input the wrong training method or you give too less input")
    else:
        pr2 =  test_x.assign(predict=pd.Series(finial_list).values)
    return pr2

def calculate_manhattan(t,p,i):
    # Manhattan Distance
    abs_subp=abs(t-p.iloc[i])
    # sum by line
    distance_list=abs_subp.sum(axis=1)
    return distance_list

def calculate_euclidean(t,p,i):
    # euclidean distance
    abs_subp=abs(t*t-p.iloc[i]*p.iloc[i])
    #sum by line
    distance_list=abs_subp.sum(axis=1)
    return distance_list

def calculate_different(t,p,i):
    #different distance
    abs_subp=abs(t-p.iloc[i])
    #sum by line
    distance_list=((abs_subp != 0).astype(int).sum(axis=1))
    return distance_list
```

There is totally three ways for measure the distance, I choose Manhattan, Euclidean and the difference of element to calculate distance

```python
# method----1: use manhattan to calculate distance  
# 2: use Euclidean to calculate distance
# 3: use the difference of element as distance
# ktime-----how many k from range to ktime we need to test
# time-----for each k, how many times should we test
# the function is used for test
def testing(method,ktime,times,if_re=False):
    ktimes=ktime+1
    # set the defalut value
    bestk=1
    
    best_correctness=0
    best_precision = 0
    best_recall = 0
    best_F_score = 0
    
    correctness_list=[]
    precision_list=[]
    recall_list=[]
    F_score_list=[]
    
    print("The value is Accuarcy, Precision, Recall, F_score resepectively")
    for k in range(1,ktimes):
        k_correctness = 0
        k_precision = 0
        k_recall = 0
        k_F_score = 0
        tmp_correctness=[]
        tmp_precision=[]
        tmp_recall=[]
        tmp_F_score=[]
        
        for i in range(times):
            train_x,train_y,test_x,test_y=separate_df(train,if_reverse=if_re)
            
            
            # use the judge_knn_m to make the prediction by different method
            df=judge_knn_m(1,method,k,train_x,train_y,test_x,test_y)
            
            
            # predict colounm
            pred=df.iloc[:,-1]
            # actual colounm
            actual=df.iloc[:,-2]
            
            # put these result to the cor_percentage to get the F-score
            tmp=cor_percentage(pred,actual)
            
            tmp_correctness.append(tmp[0])
            tmp_precision.append(tmp[1])
            tmp_recall.append(tmp[3])
            tmp_F_score.append(tmp[2])
        
        
        #get the mean of F-score
        c_tmp=sum(tmp_correctness)/times
        p_tmp=sum(tmp_precision)/times
        r_tmp=sum(tmp_recall)/times
        F_tmp=sum(tmp_F_score)/times
        
        # keep record the best one
        if(c_tmp>k_correctness):
            k_correctness=c_tmp
        if(p_tmp>k_precision):
            k_precision=p_tmp
        if(p_tmp>k_recall):
            k_recall=r_tmp
        if(F_tmp>k_F_score):
            k_F_score=F_tmp
        
        # print the current k and F-score
        print(k)
        print(k_correctness)
        print(k_precision)
        print(k_recall)
        print(k_F_score)
        
        # append the current best to the  list for make a graph
        correctness_list.append(k_correctness)
        precision_list.append(k_precision)
        recall_list.append(k_recall)
        F_score_list.append(k_F_score)

        # save the best one
        if k_correctness > best_correctness:
            best_correctness=k_correctness
        if k_precision > best_precision:
            best_precision=k_precision
        if k_recall > best_recall:
            best_recall=k_recall
        if k_F_score > best_F_score:
            bestk=k
            best_F_score=k_F_score
            
    index_list=list(range(1,ktimes))
    #make a graph
    plt.plot(index_list,correctness_list,color="red",label="Accuracy") 
    plt.plot(index_list,precision_list,color="blue",label="Precision")
    plt.plot(index_list,recall_list,color="yellow",label="Recall")
    plt.plot(index_list,F_score_list,color="green",label="F-score")
    plt.legend()
    plt.show()
    
    print("The best case for each value(bestk is base on the F_Score):")
    list_score=[bestk,best_correctness,best_precision,best_recall,best_F_score]
    print(list_score)
    return list_score
```

```python
#1: use Manhattan to calculate distance
ktime=10
time=10
Manhattan_score=testing(1,ktime,time)
```

```python
#2: use Euclidean to calculate distance 
ktime=10
time=10
Euclidean_score=testing(2,ktime,time)
```

```python
#3: use the difference of element as distance
ktime=10
time=10
Difference_score=testing(3,ktime,time)
```

```python
# Over view
method_list=['Manhattan','Euclidean','Difference']
score=['Accuarcy', 'Precision', 'Recall', 'F_score']
round_num=3
M_c=round_list(Manhattan_score[1:],round_num)
E_c=round_list(Euclidean_score[1:],round_num)
D_c=round_list(Difference_score[1:],round_num)
       
# display(Manhattan_score)
# display(Euclidean_score)
# display(Difference_score)
bar = Bar()
bar.add_xaxis(method_list)
for i in range(len(score)):
    tmp=[]
    tmp.append(M_c[i])
    tmp.append(E_c[i])
    tmp.append(D_c[i])
#     display(tmp)
#     display(score[i])
    
    # draw a graph
    name=str(score[i])
    bar.add_yaxis(name, tmp)

bar.set_global_opts(title_opts=opts.TitleOpts(title="KNN",subtitle="Best score for each method among the k"))
bar.render_notebook()
```

Thus, we can found that the first method is the best, and when k is equal to 3.
